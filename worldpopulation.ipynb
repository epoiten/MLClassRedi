How much did the Philippine population grow in the past 3 years since 2023?

```python
pop2023 = int(df.loc[df['country'] == 'Philippines', '2023 population'].values[0]) 
pop2020 = int(df.loc[df['country'] == 'Philippines', '2020 population'].values[0])
popresult = pop2023 - pop2020
print(f'The popultation increase of the Philippines for the past 3 years since 2023 is {popresult}')
```

```python
years = ['2010 population','2015 population','2020 population', '2022 population', '2023 population']

# Extract Philippines data
philippines_pop = df[df['country'] == 'Philippines'][years].values.flatten()

# Extract another country's population for comparison (e.g., India)
india_pop = df[df['country'] == 'India'][years].values.flatten()

# Plot the line chart
plt.figure(figsize=(8, 5))
plt.plot(years, philippines_pop, label='Philippines', marker='o')
plt.plot(years, india_pop, label='India', marker='o')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth Comparison: Philippines vs. India')
plt.legend()
plt.show()
```

```python
countries = df.nlargest(10, '2023 population')
philippines_population = df[df['country'] == 'Philippines']['2023 population'].values[0]

# Append Philippines if it's not in the top 10
if 'Philippines' not in countries['country'].values:
    countries = pd.concat([countries, df[df['country'] == 'Philippines']])

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(countries['2023 population'], labels=countries['country'], autopct='%1.1f%%')
plt.title('Population Share of Philippines vs Top Countries (2023)')
plt.show()
```

```python
# Plot scatter plot of population density vs. population size
plt.figure(figsize=(10, 6))
plt.scatter(df['density (km²)'], df['2023 population'], color='b', alpha=0.5)

# Highlight the Philippines
philippines_data = df[df['country'] == 'Philippines']
plt.scatter(philippines_data['density (km²)'], philippines_data['2023 population'], color='r', label='Philippines', s=100)

plt.xlabel('Population Density (people/km²)')
plt.ylabel('Population (2023)')
plt.title('Population vs. Density: Philippines vs. World')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Select countries for comparison
countries = df[df['country'].isin(['Philippines', 'India', 'China', 'USA', 'Indonesia'])]

# Set index as country and select relevant columns
heatmap_data = countries.set_index('country')[['2020 population', '2022 population', '2023 population', 'density (km²)']]

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='Blues', fmt="g")
plt.title('Population and Density Comparison of Selected Countries')
plt.show()
```

```python
# Sort the dataset to get the top 10 most populated countries
top_10_countries = df.nlargest(10, '2023 population')
if 'Philippines' not in top_10_countries['country'].values:
    philippines_data = df[df['country'] == 'Philippines']
    top_10_countries = pd.concat([top_10_countries, philippines_data])
# Create a color list where Philippines is red, others are skyblue
colors = ['red' if country == 'Philippines' else 'skyblue' for country in top_10_countries['country']]

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['country'], top_10_countries['2023 population'], color=colors)

# Add titles and labels
plt.title('Top 10 Most Populated Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
