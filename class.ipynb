```python
city_price = df.groupby('City Name')[['Low Price', 'High Price']].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='City Name', y='Low Price', data=city_price, label='Low Price', color='b')
sns.barplot(x='City Name', y='High Price', data=city_price, label='High Price', color='r', alpha=0.5)

plt.xlabel('City Name')
plt.ylabel('Price')
plt.title('Average Low and High Prices by City')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
type_counts = df['Variety'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Types')
plt.show()
```

```python
variety_price = df.groupby('Variety')[['Low Price', 'High Price']].mean().reset_index()
variety_price['Average Price'] = variety_price[['Low Price', 'High Price']].mean(axis=1)
variety_price
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Variety', y='Average Price', data=variety_price, color='orange')
plt.xlabel('Variety')
plt.ylabel('Average Price')
plt.title('Average Prices by Variety')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
